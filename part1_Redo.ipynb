# Sports Analytics - Reading Excel Files

```python
# Import libraries needed to read and work with data
import pandas as pd
import numpy as np
import os
```

```python
# Load the player data from a CSV file and show a preview
player_data = pd.read_csv('Data/player_data.csv')
print("Player Data:")
print(player_data.head())  # Show first few rows
print(f"\nShape: {player_data.shape}")  # Show how many rows and columns
print(f"Columns: {player_data.columns.tolist()}")  # Show column names
```

```python
# Load the game stint data from a CSV file and show a preview
stint_data = pd.read_csv('Data/stint_data.csv')
print("Stint Data:")
print(stint_data.head())  # Show first few rows
print(f"\nShape: {stint_data.shape}")  # Show how many rows and columns
print(f"Columns: {stint_data.columns.tolist()}")  # Show column names
```

```python
# Get final score per game
final_scores = (
    stint_data
    .groupby('game_id', as_index=False)
    .agg({
        'h_team': 'first',
        'a_team': 'first',
        'h_goals': 'max',
        'a_goals': 'max'
    })
)

# Determine winner
final_scores['winner'] = np.where(
    final_scores['h_goals'] > final_scores['a_goals'],
    final_scores['h_team'],
    np.where(
        final_scores['a_goals'] > final_scores['h_goals'],
        final_scores['a_team'],
        'Draw'
    )
)

final_scores
```

```python
player_cols_home = ['home1','home2','home3','home4']
player_cols_away = ['away1','away2','away3','away4']

def explode_stints(stints):
    rows = []
    for _, r in stints.iterrows():
        for side, cols in [('home', player_cols_home), ('away', player_cols_away)]:
            for p in cols:
                rows.append({
                    'game_id': r.game_id,
                    'player': r[p],
                    'side': side,
                    'minutes': r.minutes,
                    'team_goals': r.h_goals if side == 'home' else r.a_goals,
                    'opp_goals': r.a_goals if side == 'home' else r.h_goals,
                    'home_win': int(
                        (r.h_goals > r.a_goals) if side == 'home' else (r.a_goals > r.h_goals)
                    )
                })
    return pd.DataFrame(rows)

player_stints = explode_stints(stint_data)
player_stints
```

```python
player_stints['goal_diff_per_min'] = (
    (player_stints.team_goals - player_stints.opp_goals)
    / player_stints.minutes
)

player_stints['team_goals_per_min'] = player_stints.team_goals / player_stints.minutes
player_stints['opp_goals_per_min']  = player_stints.opp_goals  / player_stints.minutes
player_stints
```

```python
ratings = player_data.set_index('player')['rating']

def add_teammate_features(df):
    df = df.copy()
    for gid in df.game_id.unique():
        mask = df.game_id == gid
        players = df.loc[mask, 'player']
        r = ratings.loc[players].values
        df.loc[mask, 'avg_teammate_rating'] = np.mean(r)
        df.loc[mask, 'sum_teammate_rating'] = np.sum(r)
        df.loc[mask, 'max_teammate_rating'] = np.max(r)
    return df

player_stints = add_teammate_features(player_stints)
player_stints
```

```python
opp_avg_diff = (
    player_stints.groupby('game_id')['goal_diff_per_min']
    .mean()
    .rename('avg_opp_diff')
)

player_stints = player_stints.merge(opp_avg_diff, on='game_id')
player_stints
```

```python
player_stints['win_weighted_diff'] = (
    player_stints.goal_diff_per_min * (1 + player_stints.home_win)
)
player_stints
```

```python
features = [
    'team_goals_per_min',
    'opp_goals_per_min',
    'avg_teammate_rating',
    'sum_teammate_rating',
    'max_teammate_rating',
    'avg_opp_diff',
    'minutes',
    'home_win'
]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
X = player_stints[features]
y = player_stints['goal_diff_per_min']
```

```python
rf = RandomForestRegressor(
    n_estimators=100,
    max_depth=8,
    min_samples_split=10,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
)

## CV later

rf.fit(X, y)
```

```python
# Model Evaluation
y_pred = rf.predict(X)

print("R²:", r2_score(y, y_pred))
print("MSE:", mean_squared_error(y, y_pred))
print("MAE:", mean_absolute_error(y, y_pred))
```

```python
player_stints['predicted_score'] = rf.predict(X)

player_scores = (
    player_stints
    .groupby('player')['predicted_score']
    .mean()
    .reset_index()
)

player_scores
```

### Country Player Score Breakdown ML #2

```python
# Columns in stint_data that contain player names
player_cols = ['home1','home2','home3','home4','away1','away2','away3','away4']

# Step 1: Extract all unique players
all_players = pd.unique(stint_data[player_cols].values.ravel())
player_info = pd.DataFrame({'player': all_players})

# Step 2: Extract country from player name (assuming format 'CAN_12')
player_info['country'] = player_info['player'].str.split('_').str[0]

# Step 3: Merge in player ratings
# Assume you have a dataframe 'player_ratings' with columns ['player', 'rating']
player_info = player_info.merge(player_data, on='player', how='left')

# Step 4: Compute total minutes played per player
player_minutes = (
    pd.concat([
        stint_data[['home1','minutes']].rename(columns={'home1':'player'}),
        stint_data[['home2','minutes']].rename(columns={'home2':'player'}),
        stint_data[['home3','minutes']].rename(columns={'home3':'player'}),
        stint_data[['home4','minutes']].rename(columns={'home4':'player'}),
        stint_data[['away1','minutes']].rename(columns={'away1':'player'}),
        stint_data[['away2','minutes']].rename(columns={'away2':'player'}),
        stint_data[['away3','minutes']].rename(columns={'away3':'player'}),
        stint_data[['away4','minutes']].rename(columns={'away4':'player'})
    ])
    .groupby('player')['minutes']
    .sum()
    .reset_index(name='total_minutes')
)
player_info = player_info.merge(player_minutes, on='player', how='left')

# Step 5: Count number of stints per player
player_stints_count = (
    pd.concat([
        stint_data[['home1']].rename(columns={'home1':'player'}),
        stint_data[['home2']].rename(columns={'home2':'player'}),
        stint_data[['home3']].rename(columns={'home3':'player'}),
        stint_data[['home4']].rename(columns={'home4':'player'}),
        stint_data[['away1']].rename(columns={'away1':'player'}),
        stint_data[['away2']].rename(columns={'away2':'player'}),
        stint_data[['away3']].rename(columns={'away3':'player'}),
        stint_data[['away4']].rename(columns={'away4':'player'})
    ])
    .groupby('player')
    .size()
    .reset_index(name='num_stints')
)
player_info = player_info.merge(player_stints_count, on='player', how='left')

# Final player_info dataframe
print(player_info.head())

player_scores = player_scores.merge(player_info[['player','country']], on='player')
```

```python
country_tables = {}

for opp_country in player_scores.country.unique():
    temp = player_stints.merge(
        player_scores[['player','country']],
        on='player'
    )
    temp = temp[temp.country != opp_country]

    ranks = (
        temp.groupby('player')['predicted_score']
        .mean()
        .sort_values(ascending=False)
        .reset_index()
    )

    country_tables[opp_country] = ranks
```

```python
final_tables = {}

for country in player_scores.country.unique():
    df = pd.DataFrame(index=player_scores[player_scores.country==country]['player'])
    for opp in player_scores.country.unique():
        df[opp] = (
            country_tables[opp]
            .set_index('player')
            .loc[df.index, 'predicted_score']
            .rank(ascending=False)
        )
    final_tables[country] = df
```
